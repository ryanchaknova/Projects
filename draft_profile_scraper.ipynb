```python
#THE GOAL OF THIS FILE WILL BE TO BASED ON A PLAYERS NAME, GET ALL THEIR PRE-NBA INFO
#THIS FILE WILL START WITH BASKETBALL REFERENCE, BUT WITH OBJECTS I MAY MAKE MORE FUNCTIONS 

import pandas as pd
import time
```

```python
def get_player_data(player_name,index): #we must remember, players who have repeat names have an index in their url that is NOT 1, 
    #i suppose i will have to create a checker for theese with some sort of key
    #format player name for url
    player_name_url = player_name.replace(' ', '-')#replace spaces with '-'
    player_name_url=player_name_url.lower() #convert entire name to lowercase
    url=f"https://www.sports-reference.com/cbb/players/{player_name_url}-{index}.html"
    df=pd.read_html(url)#scrape url
    data=pd.DataFrame()
    
    for i in range(len(df)): #go through every table   
        df[i].drop(df[i].columns[df[i].columns.str.contains('Unnamed',case = False)],axis = 1, inplace = True) #drop the unnamed cols by name subset
        if(i==0): #season per game stats
            data=df[i] #start player df with the first table
            data.insert(0,"Player",player_name) #add players name
        elif(i==1): #conf per game stats
            df[i].drop(["Season","School","Conf","Class"], axis=1,inplace=True) #drop redundent cols
            df[i]=df[i].rename(columns={"G":"G_conf","GS":"GS_conf","MP":"MP_conf","FG":"FG_conf","FGA":"FGA_conf","FG%":"FG%_conf","2P":"2P_conf","2PA":"2PA_conf","2P%":"2P%_conf","3P":"3P_conf","3PA":"3PA_conf","3P%":"3P%_conf","FT":"FT_conf","FTA":"FTA_conf","FT%":"FT%_conf","ORB":"ORB_conf","DRB":"DRB_conf","TRB":"TRB_conf","AST":"AST_conf","STL":"STL_conf","BLK":"BLK_conf","TOV":"TOV_conf","PF":"PF_40","PTS":"PTS_40","SOS":"SOS_40"})
            data = pd.concat([data, df[i]],axis=1, sort=False) #merge into player df, all the rest are about the same
            #print(data)
        elif(i==4): #season per 40m stats, i skipped totals
            df[i].drop(["Season","School","Conf","Class","G","GS","FG%","2P%","3P%","FT%"], axis=1,inplace=True)
            df[i]=df[i].rename(columns={"MP":"MP_40","FG":"FG_40","FGA":"FGA_40","2P":"2P_40","2PA":"2PA_40","3P":"3P_40","3PA":"3PA_40","FT":"FT_40","FTA":"FTA_40","TRB":"TRB_40","AST":"AST_40","STL":"STL_40","BLK":"BLK_40","TOV":"TOV_40","PF":"PF_40","PTS":"PTS_40"})
            data = pd.concat([data, df[i]],axis=1, sort=False)
        elif(i==5): #conf per 40m stats 
            df[i].drop(["Season","School","Conf","Class","G","GS","FG%","2P%","3P%","FT%"], axis=1,inplace=True)
            df[i]=df[i].rename(columns={"MP":"MP_40_conf","FG":"FG_40_conf","FGA":"FGA_40_conf","2P":"2P_40_conf","2PA":"2PA_40_conf","3P":"3P_40_conf","3PA":"3PA_40_conf","FT":"FT_40_conf","FTA":"FTA_40_conf","TRB":"TRB_40_conf","AST":"AST_40_conf","STL":"STL_40_conf","BLK":"BLK_40_conf","TOV":"TOV_40_conf","PF":"PF_40_conf","PTS":"PTS_40_conf"})
            data = pd.concat([data, df[i]],axis=1, sort=False)
        elif(i==6): #season per 100p stats
            df[i].drop(["Season","School","Conf","Class","G","GS","FG%","2P%","3P%","FT%"], axis=1,inplace=True)
            df[i]=df[i].rename(columns={"MP":"MP_100","FG":"FG_100","FGA":"FGA_100","2P":"2P_100","2PA":"2PA_100","3P":"3P_100","3PA":"3PA_100","FT":"FT_100","FTA":"FTA_100","TRB":"TRB_100","AST":"AST_100","STL":"STL_100","BLK":"BLK_100","TOV":"TOV_100","PF":"PF_100","PTS":"PTS_100"})
            data = pd.concat([data, df[i]],axis=1, sort=False)
        elif(i==7): #conf per 100p stats
            df[i].drop(["Season","School","Conf","Class","G","GS","FG%","2P%","3P%","FT%"], axis=1,inplace=True)
            df[i]=df[i].rename(columns={"MP":"MP_100_conf","FG":"FG_100_conf","FGA":"FGA_100_conf","2P":"2P_100_conf","2PA":"2PA_100_conf","3P":"3P_100_conf","3PA":"3PA_100_conf","FT":"FT_100_conf","FTA":"FTA_100_conf","TRB":"TRB_100_conf","AST":"AST_100_conf","STL":"STL_100_conf","BLK":"BLK_100_conf","TOV":"TOV_100_conf","PF":"PF_100_conf","PTS":"PTS_100_conf","ORtg":"ORtg_conf","DRtg":"DRtg_conf"})
            data = pd.concat([data, df[i]],axis=1, sort=False)
        elif(i==8): #season adv stats
            df[i].drop(["Season","School","Conf","Class","G","GS","MP"], axis=1,inplace=True)
            data = pd.concat([data, df[i]],axis=1, sort=False)
        elif(i==9): #conf adv stats, (O,D)BPM is not provided on confrence games for some reason
            df[i].drop(["Season","School","Conf","Class","G","GS","MP","OBPM","DBPM","BPM"], axis=1,inplace=True)
            df[i]=df[i].rename(columns={"PER":"PER_conf","TS%":"TS%_conf","eFG%":"eFG%_conf","3PAr":"3PAr_conf","FTr":"FTr_conf","PProd":"PProd_conf","ORB%":"ORB%_conf","DRB%":"DRB%_conf","TRB%":"TRB%_conf","AST%":"AST%_conf","STL%":"STL%_conf","BLK%":"BLK%_conf","TOV%":"TOV%_conf","USG%":"USG%_conf","OWS":"OWS_conf","DWS":"DWS_conf","WS":"WS_conf","WS/40":"WS/40_conf"})
            data = pd.concat([data, df[i]],axis=1, sort=False)
    return data

#display(get_player_data("Anthony Bennett"))
```
